### Identify and analyze patterns in the data.

```python
num_column = DF[['TrafficIndexLive',	'JamsCount',	'JamsDelay',	'JamsLength', 'TrafficIndexWeekAgo','TravelTimeHistoric','TravelTimeLive']]
correlation_matrix= num_column.corr()
plt.figure(figsize=(9, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
#  TrafficIndexLive
plt.figure(figsize=(9, 6))
sns.histplot(DF['TrafficIndexLive'], bins=10, kde=True)
plt.title('Distribution of Traffic Index Live')
plt.xlabel('Traffic Index Live')
plt.ylabel('Frequency')
plt.show()
```

```python
#  Jamscount
plt.figure(figsize=(9, 6))
sns.histplot(DF['JamsCount'], bins=10, kde=True)
plt.title('Distribution of Jams Count')
plt.xlabel('Jams Count')
plt.ylabel('Frequency')
plt.show()
```

## 2. Data Visualization

```python
#  Jamscount  City
plt.figure(figsize=(9, 6))
sns.barplot(x='City', y='JamsCount', data=DF)
plt.title('Relationship between Jams Count and City')
plt.xlabel('City')
plt.ylabel('Jams Count')
plt.show()
```

```python
#       
plt.figure(figsize=(9, 6))
sns.scatterplot(x='TrafficIndexLive', y='JamsDelay', data=DF)
plt.title('Relationship between Traffic Index Live and Jams Delay')
plt.xlabel('Traffic Index Live')
plt.ylabel('Jams Delay')
plt.show()
```

```python
#   
city_jams = DF.groupby('City')['TrafficIndexLive'].sum()
plt.figure(figsize=(10, 8))
plt.pie(city_jams, labels=city_jams.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Jams in Each City')
plt.show()
```
