```python
import numpy as np
# تعديل أنواع البيانات
# Time & Driving_experience & Age_band_of_driver & Number_of_vehicles_involved & Service_year_of_vehicle & Number_of_casualties
Df['Time'] = pd.to_datetime(Df['Time'], format='%H:%M:%S', errors='coerce').dt.time

def convert_range_to_mean(value): #عشان اخذ المتوسط لها وبعدين نحذف -
    if pd.isna(value) or value == 0:
        return value
    if isinstance(value, str) and '-' in value:
        try:
            start, end = map(float, value.split('-'))
            return (start + end) / 2
        except ValueError:
            return np.nan
    try:
        return float(value)
    except ValueError:
        return np.nan
for column in columns_to_convert:
  Df[columns_to_convert] = Df[columns_to_convert].replace('', 0)
  Df[column] = Df[column].apply(convert_range_to_mean).astype('float')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
num_column = Df[['Age_band_of_driver',	'Driving_experience',	'Service_year_of_vehicle',	'Number_of_vehicles_involved', 'Number_of_casualties']]
correlation_matrix= num_column.corr()
plt.figure(figsize=(9, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Visualize the distribution of key variables (e.g., Age_band_of_driver, Type_of_vehicle).

```python
plt.figure(figsize=(9, 6))
sns.histplot(Df['Age_band_of_driver'], kde=True, bins=30)
plt.title('Distribution of Age_band_of_driver')
plt.xlabel('Age_band_of_driver')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

### Explore relationships between variables (e.g., Age_band_of_driver vs. Accident_severity).


```python
 # العلاقة بين Age_band_of_driver و Accident_severity
age_accident_severity = Df.groupby('Age_band_of_driver')['Accident_severity'].value_counts(normalize=True).unstack().fillna(0)
age_accident_severity.plot(kind='bar', stacked=True, figsize=(9, 7))
plt.title('Age Band of Driver vs. Accident Severity')
plt.xlabel('Age Band of Driver')
plt.ylabel('%')
plt.legend(title='Accident Severity')
plt.show()
```
