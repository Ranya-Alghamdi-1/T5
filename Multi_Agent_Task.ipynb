### Step 3: Create Agents

Now, define the two agents. **Agent 1** will handle research, and **Agent 2** will summarize the research.

Fill in the code below to create the agents:


```python
llm = ChatGroq(
    model="groq/llama-3.1-70b-versatile",
    temperature=0.7,
    max_tokens=None,
    timeout=None,
    max_retries=2
)


researcher = Agent(
    llm=llm,
    role="Researcher",
    goal="Write a research on a specified topic and provide detailed information.",
    backstory="You are an expert researcher with the ability to gather and organize relevant information.",
    allow_delegation=False,
    verbose=1,
)

summarizer = Agent(
    llm=llm,
    role="Summarizer",
    goal="Summarize the research conducted by Researcher Agent into a concise, easy-to-read format.",
    backstory="You are a skilled summarizer with expertise in transforming complex information into clear, concise summaries.",
    allow_delegation=False,
    verbose=1
)
```


---
### Step 4: Execute the Multi-Agent System

Finally, run the system to allow the agents to collaborate and complete their tasks.


```python
task1 = Task(
    description="Conduct research on a given topic and write a detailed report.",
    expected_output="A detailed research report on the specified topic.",
    output_file="research_report.txt",
    agent=researcher
)

task2 = Task(
    description="Summarize the research report for a broader audience.",
    expected_output="A concise summary of the research report.",
    output_file="research_summary.txt",
    agent=summarizer,
    input_file="research_report.txt"
)
```

```python
crew = Crew(agents=[researcher, summarizer], tasks=[task1, task2], verbose=1)

print(crew.kickoff())
```
